5.Используя pandas, проведи когортный анализ пользователей. В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц

```python
# соединяем product_id и даты покупок. Создаю условие только для оплаченных товаров
df_cohort = df_orders.merge(df_customers, how='left', on = 'customer_id')\
    .query('order_approved_at != "NaT" & order_status not in ["canceled", "unavailable"]')

# указываю лишних клиентов
out_customer = df_cohort.query('order_approved_at < "2017-01-01" or order_approved_at >= "2018-01-01"').customer_unique_id

# меняем формат даты на месяца
df_cohort.order_approved_at = df_cohort.order_approved_at.dt.to_period("M")

# вычитываем лишних клиентов
df_cohort =df_cohort.query('customer_unique_id not in @out_customer').sort_values('order_approved_at')

# узнаем стоимоть заказа
df_order_price = df_orders_items.groupby('order_id', as_index = False).agg({'price' : 'sum'})

# крепим цену
df_cohort = df_cohort.merge(df_order_price, how='left', on = 'order_id')
```

```python
# находим дату первой покупки каждого клиента
df_first_buy = df_cohort.groupby('customer_unique_id', as_index = False)\
    .agg({'order_approved_at' : 'min'})\
    .sort_values('order_approved_at')\
    .rename(columns={"order_approved_at": "date_first_buy"})

# Крепим дату первой покупки каждого клиента к общей таблице
df_cohort = df_cohort.merge(df_first_buy, how='left', on = 'customer_unique_id')

# Поулчаем Когорты
df_cohort = df_cohort.groupby(['date_first_buy', 'order_approved_at'], as_index = False )\
                    .agg({'order_id' :'count', 'price' : 'sum'})\
                    .rename(columns={"order_id": "count", "price": "sum"})
df_cohort
```

```python
# Сортируем количество заказов на 3 месяц по возрастанию
df_cohort.query('(order_approved_at - date_first_buy) == @MonthEnd(2)').sort_values(['count', 'sum'])
```

6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю. 

```python
# собираем данные в одну таблицу
df_RFM = df_orders.merge(df_customers, how='left', on = 'customer_id')\
    .query('order_approved_at != "NaT" & order_status not in ["canceled", "unavailable"]')

# крепим цену
df_RFM = df_RFM.merge(df_order_price, how='left', on = 'order_id')
```

```python
# дата начала и конца данных
print('Orders from {} to {}'.format(df_RFM['order_approved_at'].min(),
                                    df_RFM['order_approved_at'].max()))
```

```python
# отчетную дату ставлю следующим днем полсе окончания данных
now = df_RFM['order_approved_at'].max() + dt.timedelta(days=1)

# отчетный период беру - 365 последних дней ( последний год)
period = 365
```

```python
# добавляем столбец указывающий количество времени от отчетной даты до даты покупки
df_RFM['date_diff'] = df_RFM['order_approved_at'].apply(lambda x: (now - x).days)
```

```python
#рассчитываем количество повторений за последний год и количество времени для каждого клиента
RFM = df_RFM.groupby('customer_unique_id', as_index = False). \
    agg({'date_diff': 'min','order_approved_at': lambda x: len([d for d in x if d >= now - dt.timedelta(days=period)]) })

# меняем название столбцов
RFM.rename(columns={'date_diff': 'Recency', 'order_approved_at': 'Frequency'}, inplace=True)
```

```python
# задаю конкретную дату начиная с которой анализируем данные
diff = now - dt.timedelta(days = period)

# счиатю стоимость заказов, выполненных одним клиентом
df_price = df_RFM.query('order_approved_at > @diff').groupby('customer_unique_id', as_index = False).agg({'price' : 'sum'})

# прикрепляю стоимость к общему отчету
RFM = RFM.merge(df_price, how='left', on = 'customer_unique_id').rename(columns={'price': 'Monetary'})
```

```python
# убираю данные вышедшие за исследуемый период 
RFM = RFM.query('Frequency > 0')
RFM
```

```python
# Находим границы метрик с помощью квинтилей (квантиль по равным долям)
quantiles = RFM.quantile(q=[0.2, 0.4, 0.6, 0.8])
quantiles = quantiles.to_dict()

#готовим итоговую таблицу с сегментами
RFM_segment = RFM.copy()
```

```python
# Создаем функцию для сегментации R (время от последней покупки) оценка идет от обратного
def R_Class(x,p,d):
    if x <= d[p][0.2]:
        return 5
    elif x <= d[p][0.4]:
        return 4
    elif x <= d[p][0.6]: 
        return 3
    elif x <= d[p][0.8]: 
        return 2
    else:
        return 1

# Создаем функцию для сегментации F и M    
def FM_Class(x,p,d):
    if x <= d[p][0.2]:
        return 1
    elif x <= d[p][0.4]:
        return 2
    elif x <= d[p][0.6]: 
        return 3
    elif x <= d[p][0.8]: 
        return 4
    else:
        return 5    
```

```python
RFM_segment['R_Quartile'] = RFM_segment['Recency'].apply(R_Class, args=('Recency',quantiles,))
RFM_segment['F_Quartile'] = RFM_segment['Frequency'].apply(FM_Class, args=('Frequency',quantiles,))
RFM_segment['M_Quartile'] = RFM_segment['Monetary'].apply(FM_Class, args=('Monetary',quantiles,))
RFM_segment['RFMClass'] = RFM_segment.R_Quartile.map(str) \
                            + RFM_segment.F_Quartile.map(str) \
                            + RFM_segment.M_Quartile.map(str)
```

```python
# получаем таблицу с сегментацией клиентов, лучшая оценка это "5" за каждый раздел (Quartile). 
# RFMClass - набор оценок  -лучшее сочетание трех групп - "555"
RFM_segment
```
